### Funcion de activación

```python
def activacion(x):
    return 1/(1+np.exp(-x))
```

### Forward propagation

<h3 align = 'center'> Variables </h3>

|Codigo | Math Symbol | Definition 
| :-: | :-: | :-: |
|X|$$X$$|Datos de entrada| 
|y |$$y$$|Datos target|
|W1 | $$W^{(1)}$$ | Pesos Capa1 | 
|W2 | $$W^{(2)}$$ | Pesos Capa 2 | 
|z2 | $$z^{(2)}$$ | Activación capa 2 | 
|a2 | $$a^{(2)}$$ | Actividad post funcion f(z2) | 
|z3 | $$z^{(3)}$$ | Activación capa 3 |


La formula del feedforward establece una secuencia de pasos que va propagando a través de sucesivas multiplicaciones de matrices el ajuste necesario el W1 y W2 que se pueden asimilar dentro del marco de un modelo a la sinapsis neuronal. El modelo feedforward se resumen matemáticamente en 4  ecuaciones que esquematizan el impacto de cada uno de los ejemplos del dataset(en un escenario de gradiente decente de tipo batch). Podemos ver estas ecuaciones a continuación:


$$
z^{(2)} = XW^{(1)} \tag{1}\\
$$
$$
a^{(2)} = f(z^{(2)}) \tag{2}\\
$$
$$
z^{(3)} = a^{(2)}W^{(2)} \tag{3}\\
$$
$$
\hat{y} = f(z^{(3)}) \tag{4}\\
$$

```python
def forward(X):
    z2 = np.dot(X, W1)  #(1)
    a2 = activacion(z2) #(2)
    z3 = np.dot(a2, W2) #(3)
    y_ = activacion(z3) #(4)
    return y_
```

```python
Y_ = forward(X)
Y_
```
